```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
#Scrape pages from College Football Reference website
years = list(range(1999, 2022))
for year in years:
  pass_page = requests.get(f'https://www.sports-reference.com/cfb/years/{year}-passing.html')
  with open('/content/drive/MyDrive/Project Notebooks/scraped pages/Combine Pages/cfbpassing{}.html'.format(year), 'w+') as f1:
    f1.write(pass_page.text)
  rush_page = requests.get(f'https://www.sports-reference.com/cfb/years/{year}-rushing.html')
  with open('/content/drive/MyDrive/Project Notebooks/scraped pages/Combine Pages/cfbrushing{}.html'.format(year), 'w+') as f2:
    f2.write(rush_page.text)
  rec_page = requests.get(f'https://www.sports-reference.com/cfb/years/{year}-receiving.html')
  with open('/content/drive/MyDrive/Project Notebooks/scraped pages/Combine Pages/cfbreceiving{}.html'.format(year), 'w+') as f3:
    f3.write(rec_page.text)
```

```python
#Parse and combine pages into dataframes
rec_table = []
rush_table = []
pass_table = []

for year in years:
  with open('/content/drive/MyDrive/Project Notebooks/scraped pages/Combine Pages/cfbpassing{}.html'.format(year)) as f1:
    pass_page = f1.read()
    pass_soup = BeautifulSoup(pass_page, 'html.parser')
    table1 = pass_soup.find(id='passing')
    pass_soup.find('tr', class_ = 'over_header').decompose()
    table1_read = pd.read_html(str(table1))[0]
    table1_read['Season'] = year
    pass_table.append(table1_read)
  with open('/content/drive/MyDrive/Project Notebooks/scraped pages/Combine Pages/cfbrushing{}.html'.format(year)) as f2:
    rush_page = f2.read()
    rush_soup = BeautifulSoup(rush_page, 'html.parser')
    table2 = rush_soup.find(id='rushing')
    rush_soup.find('tr', class_ = 'over_header').decompose()
    table2_read = pd.read_html(str(table2))[0]
    table2_read['Season'] = year
    rush_table.append(table2_read)
  with open('/content/drive/MyDrive/Project Notebooks/scraped pages/Combine Pages/cfbreceiving{}.html'.format(year)) as f3:
    rec_page = f3.read()
    rec_soup = BeautifulSoup(rec_page, 'html.parser')
    table3 = rec_soup.find(id='receiving')
    rec_soup.find('tr', class_ = 'over_header').decompose()
    table3_read = pd.read_html(str(table3))[0]
    table3_read['Season'] = year
    rec_table.append(table3_read)

cfb_pass = pd.concat(pass_table)
cfb_rush = pd.concat(rush_table)
cfb_rec = pd.concat(rec_table)
```

```python
#Duplicated rows are all additional header rows. Will remove.
cfb_pass.drop_duplicates(keep=False, inplace=True)
cfb_rec.drop_duplicates(keep=False, inplace=True)
cfb_rush.drop_duplicates(keep=False, inplace=True)
cfb_pass.duplicated().sum()
```

```python
#Create dictionary for iterations
dfs = {'Passing DF': cfb_pass,
       'Receiving DF': cfb_rec,
       'Rushing DF': cfb_rush}
```

```python
#Checking for NaNs

for h, i in dfs.items():
  i_report = pd.DataFrame({'Column': i.columns,
                           'Dtype': i.dtypes,
                           'null': i.isna().sum(),
                           '% null': i.isna().sum()/len(i)})
  print('Overview:', h)
  print('***'*20)
  display(i_report)
  print(h, 'Numeric Analysis:')
  print('***'*20)
  display(i.describe())
```

```python
for name, i in dfs.items():
  print(name)
  display(round(i.isna().sum()/len(i), 2))
```

No nulls in passing DF.

RB and Rec nulls are from players who did not accumulate stats in the categories that don't 100% map to their positions (receiving for RBs, rushing for Rec's). However, these are important to the analysis of each position, so I will impute 0s to these columns.

```python
cfb_rush.fillna(0, inplace=True)
cfb_rec.fillna(0,inplace=True)
for name, i in dfs.items():
  print(name)
  display(round(i.isna().sum()/len(i), 2))
```

```python
#Changing Dtypes for numeric categories
for key, value in dfs.items():
  print(key)
  display(value.columns)
```

```python
cfb_pass[['G', 'Cmp', 'Att', 'Pct', 'Yds',
       'Y/A', 'AY/A', 'TD', 'Int', 'Rate', 'Att.1', 'Yds.1', 'Avg', 'TD.1']] = cfb_pass[['G', 'Cmp', 'Att', 'Pct', 'Yds',
       'Y/A', 'AY/A', 'TD', 'Int', 'Rate', 'Att.1', 'Yds.1', 'Avg', 'TD.1']].astype(float)
cfb_rec[['G', 'Rec', 'Yds', 'Avg', 'TD', 'Att',
       'Yds.1', 'Avg.1', 'TD.1', 'Plays', 'Yds.2', 'Avg.2', 'TD.2']] = cfb_rec[['G', 'Rec', 'Yds', 'Avg', 'TD', 'Att',
       'Yds.1', 'Avg.1', 'TD.1', 'Plays', 'Yds.2', 'Avg.2', 'TD.2']].astype(float)
cfb_rush[['G', 'Att', 'Yds', 'Avg', 'TD', 'Rec',
       'Yds.1', 'Avg.1', 'TD.1', 'Plays', 'Yds.2', 'Avg.2', 'TD.2']] = cfb_rush[['G', 'Att', 'Yds', 'Avg', 'TD', 'Rec',
       'Yds.1', 'Avg.1', 'TD.1', 'Plays', 'Yds.2', 'Avg.2', 'TD.2']].astype(float)
for key, value in dfs.items():
  print(key)
  print(value.dtypes)
```

```python
#Remove asterisk from player names column
cfb_pass['Player'] = cfb_pass['Player'].str.replace('*', '', regex=False)
cfb_rec['Player'] = cfb_rec['Player'].str.replace('*', '', regex=False)
cfb_rush['Player'] = cfb_rush['Player'].str.replace('*', '', regex=False)
```

```python
#One last check of the DFs before export
for title, data in dfs.items():
  column_names = data.columns
  for col in column_names:
    if data[col].dtypes.name == 'object':
      print(title)
      print('\n\n')
      print(data[col].unique())
      print('\n\n')
    else:
      print(title)
      print('\n\n')
      print(data[col].describe())
      print('\n\n')
```

```python
#Aggregate player's seasonal data to produce single career average
pass_pivot = cfb_pass.pivot_table(columns='Player', aggfunc = 'mean')
rush_pivot = cfb_rush.pivot_table(columns='Player', aggfunc = 'mean')
rec_pivot = cfb_rec.pivot_table(columns='Player', aggfunc = 'mean')


display(pass_pivot)
display(rush_pivot)
rec_pivot  
```

```python
#Transpose columns and index to complete aggregation
rec_pivot = rec_pivot.T.reset_index()
rush_pivot = rush_pivot.T.reset_index()
pass_pivot = pass_pivot.T.reset_index()
display(rec_pivot.head())
display(rush_pivot.head())
pass_pivot.head()
```

```python
#Drop unnecessary rows and export to Excel
rec_done = rec_pivot[['Player', 'Avg', 'Avg.2', 'TD.2', 'Plays']]
rush_done = rush_pivot[['Player', 'Avg', 'Avg.1', 'TD.2', 'Plays']]

pass_pivot['Total TDs/season'] = pass_pivot['TD'] + pass_pivot['TD.1']
pass_done = pass_pivot[['Player', 'Y/A', 'AY/A', 'Rate', 'Total TDs/season', 'Avg']]


rec_done = rec_done.rename(columns = {'Avg': 'Avg/Rec',
                           'Avg.2': 'Avg/Rush',
                           'TD.2': 'Total TDs/season',
                           'Plays': 'Total Scrimmage Plays/season'})
rush_done = rush_done.rename(columns = {'Avg': 'Avg/Rush',
                           'Avg.1': 'Avg/Rec',
                           'TD.2': 'Total TDs/season',
                           'Plays': 'Total Scrimmage Plays/season'})
pass_done = pass_done.rename(columns = {'Avg': 'Rush Avg.'})
display(rec_done.head())
display(rush_done.head())
pass_done.head()
```
